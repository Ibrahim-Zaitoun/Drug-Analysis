```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set()
```

```python
data= pd.read_csv('D:\Data\data-visualization\drug200.csv')
data.head()
```

```python
data.info() # BP -> Blood Pressure Levels (BP) , Na_to_K -> Na to Potassium Ration
```

```python
data.describe()
```

```python
sns.boxplot(data= data, x='Age')
plt.show()
```

```python
sns.lineplot(data= data, x='Age', y='Na_to_K')
plt.show()
```

```python
sns.boxplot(data= data, x='Age', y='BP')
plt.show()
```

```python
sns.countplot(data= data, x='BP', hue= 'Sex',order=['LOW','NORMAL','HIGH'], palette={'M':'blue','F':'#808080'})
plt.xlabel('Blood Pressure Levels (BP)')
plt.show()
```

```python
sns.countplot(data= data, x='Cholesterol',hue= 'Sex',hue_order=['M','F'])
plt.show()
```

```python
sns.pointplot(data= data, x= 'Age', y= 'Cholesterol')
plt.show()
```

```python
sns.countplot(data= data, x='Drug',hue= 'Sex',order=['drugA', 'drugB', 'drugC', 'drugX', 'DrugY'])
plt.show()
```

```python
sns.boxplot(data= data, x='Age',y= 'Drug',order=['drugA', 'drugB', 'drugC', 'drugX', 'DrugY'])
plt.show()
```

```python
sns.scatterplot(data= data, x='Drug',y= 'Na_to_K')
plt.show()
```

```python
sns.scatterplot(data= data, x='Drug',y= 'Age')
plt.show()
```
